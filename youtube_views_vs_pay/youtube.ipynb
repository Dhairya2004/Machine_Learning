```
import numpy as np
import pandas
import matplotlib.pyplot as plt
from grapher import Grapher
```

```
df = pandas.read_csv('ytmoney(1).csv')
x = np.array(df['views'])/100000
y = np.array(df['money0'])/10000
```

```
class Linear_model:

    def __init__(self, X, Y):
        self.x = X
        self.y = Y
        self.m = 1
        self.c = 0
        self.alpha=0.00000001
        self.iteration=200
    
    def hypo(self,x):
        return self.m*x + self.c

    def J(self, m, c):
        diff = [(self.y[i]-self.hypo(x))**2 for i, x in enumerate(self.x)]
        return (1/2)*sum(diff)

    def gradient_decent(self):
        for i in range(self.iteration):
            self.c = self.c - self.alpha*sum([(self.hypo(x) - self.y[i]) for i, x in enumerate(self.x)])
            self.m = self.m - self.alpha*sum([(self.hypo(x) - self.y[i])*x for i, x in enumerate(self.x)])
            
            # plt.plot(self.x, self.m*self.x + self.c, c="r")

        return self.m, self.c

n = 10000

model = Linear_model(x[:n], y[:n])

m, c = model.gradient_decent()


plt.scatter(x[:n], y[:n])
plt.plot(x[:n], m*x[:n] + c, c='y')
```
